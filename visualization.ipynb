# ploting multiple graphs

### Analyze the relationship between saleprice and all SF related columns

```python
sf_col=[col_name for col_name in housing.columns if "SF" in col_name]
```

```python
len(sf_col)
```

```python
fig,axs=plt.subplots(nrows=3,ncols=3,figsize=(10,10),sharey=True)
for i in range(0,len(sf_col)):
    rows=i // 3
    cols=i % 3
    ax=axs[rows,cols]
    plot=sns.regplot(x=sf_col[i],y="SalePrice",data=housing,ax=ax)
```
